```python
df['Price in thousands']=pd.to_numeric(df['Price in thousands'],errors='coerce')
df['Price in thousands']=df['Price in thousands'].fillna(df['Price in thousands'].mean())
df
```

```python
#find min and max
def find_min_max(df,col_name):
    min_value=df[col_name].min()
    max_value=df[col_name].max()
    return min_value,max_value
```

```python
min_value,max_value=find_min_max(df,'Price in thousands')
min_value
```
