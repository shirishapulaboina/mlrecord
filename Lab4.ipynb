```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
cols=df.select_dtypes(include=['category','object'])
for col in cols:
    df[col]=le.fit_transform(df[col])
df
```

```python
df
```

```python
df=pd.get_dummies(df,drop_first=True)
```

```python
df
```

```python
#apply feature scaling on numerical data
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
numericcols=df.select_dtypes(include=['int','int64','float64']).columns
df[numericcols]=ss.fit_transform(df[numericcols])
df
```

```python
y=df['price']
x=df.drop(columns=['price'])
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,adjusted_rand_score
```

```python
y_pred=model.predict(x_test)
```

```python
print(r2_score(y_test,y_pred))
```

```python
y_pred=model.predict(x_train)
error=y_pred-y_train
sns.residplot(x=y_pred,y=error)
plt.show()
```

```python
y_pred=model.predict(x_test)
error=y_pred-y_test
sns.residplot(x=y_pred,y=error)
plt.show()
```
